```python
import pandas as pd
from nlptoolkits.OpenAIKits import ClassifierT
```

```python
def cot_generator(prp, identifier_col, target_col, stop_text):
    return [
        {
            "role": "system",
            "content": f"""
                    ## -------------------------------------------------------------
                    Understanding Task Assistance Levels with OpenAI's Large Language Model (LLM)
                    As a research assistant, your role involves labeling tasks based on the degree to which
                    they can be enhanced or expedited by OpenAI's LLM. The tasks are categorized based on
                    whether the LLM can directly address them, if LLM-powered applications are required,
                    or if image capabilities are necessary.

                    Meanwhile, you could only export(return) Valid JSON data under the following guidance.

                    Below is the rubric for classifying these tasks:

                    ## -------------------------------------------------------------
                    Exposure Rubric:
                    E1 - Direct Exposure:
                    Tasks fall under this category if utilizing the LLM through interfaces
                    like ChatGPT or OpenAI Playground can reduce completion time by at least half,
                    maintaining equivalent quality.
                    Such tasks can be simplified into:

                    <E1 tasks examples start here>
                    Writing/transforming text or code per complex instructions.
                    Editing existing text or code based on specifications.
                    Coding to automate manual tasks.
                    Translating languages.
                    Summarizing medium-length documents.
                    Providing document feedback.
                    Answering document-related queries.
                    Generating relevant questions about a document.
                    Crafting interview or assessment questions.
                    Handling email correspondences, including negotiations (if written).
                    Maintaining written records.
                    Preparing general knowledge-based training materials.
                    Informing via any written or spoken medium.
                    <E1 tasks examples end here>

                    ## -------------------------------------------------------------
                    E2 - Exposure by LLM-Powered Applications:
                    Tasks require additional software built on LLM,
                    as LLM alone wouldn't cut the completion time by half.

                    This software might have features for:
                    <E2 features start here>
                    Summarizing extensive documents (>2000 words) and answering related questions.
                    Retrieving current facts online for use with LLM.
                    Searching and retrieving information from an organization’s resources.
                    Accessing specialized domain knowledge.
                    Making recommendations based on data or written input.
                    Analyzing written content for decision-making.
                    Preparing training materials with specialized knowledge.
                    Advising on various issues.
                    Managing complex databases.
                    <E2 features end here>

                    ## -------------------------------------------------------------
                    E3 - Exposure Given Image Capabilities:
                    Tasks are significantly expedited with access to LLM and a system with image capabilities
                    (viewing, captioning, creating—excluding video).
                    Limitations apply to detailed information retrieval from images. Examples include:

                    <E3 example start here>
                    Extracting text from PDFs.
                    Image scanning.
                    Creating/editing digital images per instructions (realistic but not highly detailed).
                    Object identification within images (excluding relational identification).
                    <E3 example end here>

                    ## -------------------------------------------------------------
                    E0 - No Exposure:
                    Briefly, In such work, LLM could not essentially enhance the productivity.
                    Classify tasks as E0 if the tools mentioned don't clearly reduce the time for an experienced worker
                    to complete the task by at least half. Characteristics of such tasks include:

                    <E0 example start here>
                    High necessity for human interaction (e.g., in-person demonstrations).
                    Need for precise measurements.
                    Detailed visual reviews.
                    Physical activity (use of hands, walking).
                    Decisions impacting human livelihood (LLM can recommend but not decide).
                    Legal restrictions requiring human intervention.
                    No significant time saving using LLM or LLM-based tools compared to existing technology.
                    Default category in cases of uncertainty.
                    <E0 example end here>

                    ## -------------------------------------------------------------
                    Guidelines for Usage:
                    In your role, consider the above rubric critically.
                     While the LLM can greatly assist in various tasks,
                      it is vital to recognize its limitations and legal or ethical boundaries.
                      Always prioritize tasks where the LLM's use respects these parameters
                       and truly augments productivity and efficiency.
                    """
        },
        {
            "role": "user",
            "content": f"""
                Read the following occupation title and one of the tasks that belong to that occupation.

                 Then do three things.
                  Question1: Reason step by step to decide which of the labels (E0/E1/E2/E3)
                  from the exposure rubric you were given applies to the task's exposure to LLM.
                  Report. Give an explanation for which label applies and report the label
                  that you think fits best. Do not say zero or N/A.

                  Please kindly give me your chain of though which could Step-By-Step show
                  How could you get your result, your thinking and your logic.

                  Anwser Format:
                  Free sentence(s), Never Zero or N/A NA return back

                  Question2:  Report only the label that you determined for the task,
                  which should match the label in step 1. Do not reply N/A.
                  Choices are below:
                  ['E1', 'E2', 'E3', 'E0']
                   Besides must return a string, Never NA or other format

                  Question3:  Given the amount of speculation required in step 1,
                   describe your certainty about the estimate. either high, moderate, or low.
                   Choices are below:
                   ['high', 'moderate', 'low']
                   Besides must return a string, Never NA or other format


                Your answer should return in such format:
                [
                    {{

                        'Question1':'<The anwser you given to anwser the Question1>',
                        'Question2':'<The anwser you given to anwser the Question2>',
                        'Question3':'<The anwser you given to anwser the Question3>',
                        {identifier_col}:<The origin {identifier_col} col value>

                    }}


                ]{stop_text}


                The datas are given and will always be given in such format,
                each of the sub-dict contains a whole task for you to do, you have to do these one-by-one and return the result in format you seen:
                [
                    {{
                        "{target_col}":
                        "Occupation: Inspectors, Testers, Sorters, Samplers, and Weighers.
                        Task: Adjust, clean, or repair products or processing equipment to correct defects found during inspections."

                        "{identifier_col}": "00-0000.00#STICKER#0000"

                    }}

                ]
                """
        },
        {
            "role": "assistant",
            "content": f"""
                [
                    {{
                        "Question1": "The model does not have access to any kind of physicality,
                         and more than half of the task (adjusting, cleaning and repairing equipment)
                         described requires hands or other embodiment. This kind of word neither enhance the
                         the person's productivity by LLM itself, nor enhance the person's productivity combined with other supports or methods
                         like applications or Image-Capability.
                         So E1, E2, E3 could not delegated this task.
                         it is clearly a kind of E0 works, Therefore, it falls under the category of E0.",
                        "Question2":"E0",
                        "Question3":"high",
                        "{identifier_col}": "00-0000.00#STICKER#0000"
                    }}
                ]
                """
        },
        {
            "role": "user",
            "content": f"""
                Then it is time for you to do the next question by yourself.
                And here are a task-statement combination for you to classify,

                Clearly, Here are {identifier_col} and statements:,
                The data are always, also given in list of dict format:

                {prp}
                """
        },
    ]
```

```python
sample_df = pd.read_excel('./input_data/Task_Statements.xlsx', engine='openpyxl')

sample_df['statement'] = "occupation: " + sample_df['Title'] + "\n, Task: " + sample_df['Task']
sample_df['id'] = sample_df.apply(lambda x: str(x['O*NET-SOC Code']) + '#STICKER#' + str(x['Task ID']), axis=1)


APIKEY = input('input_your_own apikey')

SAMPLES = 10
```

```python
subsample_df = sample_df[['statement', 'id']].sample(n=SAMPLES, random_state=42)
ClassifierT.chatcompletion_worker(
    prompt_generator_func=lambda dict_list: dict_list,
    statement_df=subsample_df,
    stop_text='##STOP##',
    target_col='statement',
    identifier_col='id',
    openai_apikey=APIKEY,
    chunksize=1,  # you do only could send one chunksize
    model='gpt-3.5-turbo',
    result_type='df',
    dataframe_format_error='skip',
    dataframe_deficiency_error='warn',
    is_dataframe_errors_onebyone_retry=False,
    message_generate_func=lambda prp: cot_generator(prp, 'id', 'statement', '##STOP##'),
    ratelimit_call=1000,
    ratelimit_period=60,
    backoff_max_tries=9,
)
```
